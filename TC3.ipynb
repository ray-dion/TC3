```python
# Range checks

negative = df[df['transaction_amount']<0]
if not negative.empty:
  print(negative)

negative_promo = df[df['promo_amount']<0]
if not negative_promo.empty:
  print(negative_promo)

invalid_merchant_rating = df[(df['merchant_rating']<0) | (df['merchant_rating']>5)]
if not invalid_merchant_rating.empty:
  print(invalid_merchant_rating)
```

```python
# improbable promo

improbable_promo = df[df['promo_amount']>df['transaction_amount']]
if not improbable_promo.empty:
  print(improbable_promo)
```
